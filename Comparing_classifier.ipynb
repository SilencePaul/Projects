```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Import Data

```python
data = pd.read_csv('./data/fetal_health.csv')
data.head()
```

Check data information

```python
data.info()
```

```python
data.describe().T
```

Change Fetal_health from 3 Classifications to 2 Classifications
From:
1 means Normal
2 means Suspect
3 means Pathologic
To:
0 means Normal
1 means Abnormal

```python
df = data.copy()
df['fetal_health'] = df['fetal_health'].apply(lambda x: 0 if x == 1 else 1)
df.describe().T
```

Some visualization of the data

```python
sns.countplot(x='fetal_health', data=df)
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = df.drop('fetal_health', axis=1)
target = df['fetal_health']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

pca = PCA(n_components=2)
principle_components = pca.fit_transform(scaled_features)

pca_df = pd.DataFrame(data=principle_components, columns=['PC1', 'PC2'])
pca_df['fetal_health'] = target.values

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='fetal_health', data=pca_df)
plt.show()
```

Feature Selection

```python
corr_matrix = df.corr()

plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
fetal_health_corr = corr_matrix['fetal_health'].sort_values(ascending=False)
fetal_health_corr
```

Significant features: absolute value is larger than 0.1

```python
significant_features = fetal_health_corr[abs(fetal_health_corr) > 0.1]
significant_features
```

Reframe the dataset

```python
reframed_df = df[significant_features.index]
reframed_df.head()
```

Scale reframed dataset

```python
X = reframed_df.drop('fetal_health', axis=1)
y = reframed_df['fetal_health']

from sklearn.model_selection import train_test_split
cols = list(X.columns)
scaler = StandardScaler()
X_scalered = scaler.fit_transform(X)
X_scalered = pd.DataFrame(X_scalered, columns=cols)
X_scalered.describe().T
```

Seperate test and training datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X_scalered, y, test_size=0.25, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

Select Several models and use cross validation to have a score

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

log_reg = LogisticRegression()
svc = SVC(kernel='rbf')
decision_tree = DecisionTreeClassifier()
knn = KNeighborsClassifier()
random_forest = RandomForestClassifier()

log_reg_scores = cross_val_score(log_reg, X_train, y_train, cv=10)
svc_scores = cross_val_score(svc, X_train, y_train, cv=10)
decision_tree_scores = cross_val_score(decision_tree, X_train, y_train, cv=10)
knn_scores = cross_val_score(knn, X_train, y_train, cv=10)
random_forest_scores = cross_val_score(random_forest, X_train, y_train, cv=10)

print("Logistic Regression Accuracy: %0.2f (+/- %0.2f)" % (log_reg_scores.mean(), log_reg_scores.std() * 2))
print("SVC Accuracy: %0.2f (+/- %0.2f)" % (svc_scores.mean(), svc_scores.std() * 2))
print("Decision Tree Accuracy: %0.2f (+/- %0.2f)" % (decision_tree_scores.mean(), decision_tree_scores.std() * 2))
print("KNN Accuracy: %0.2f (+/- %0.2f)" % (knn_scores.mean(), knn_scores.std() * 2))
print("Random Forest Accuracy: %0.2f (+/- %0.2f)" % (random_forest_scores.mean(), random_forest_scores.std() * 2))
```

Computation Time for each model

```python
import time

training_times = []
testing_times = []

for clf in classifiers:
    start_time = time.time()
    clf.fit(X_train, y_train)
    training_times.append(time.time() - start_time)
    
    start_time = time.time()
    clf.predict(X_test)
    testing_times.append(time.time() - start_time)

for i, clf_name in enumerate(classifier_names):
    print(clf_name)
    train_time_ms = training_times[i]*1000
    test_time_ms = testing_times[i]*1000
    print("Training time: %0.1f ms" % train_time_ms)
    print("Testing time: %0.1f ms" % test_time_ms)
    print("\n")
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(15, 12))
labels = ['Normal', 'Abnormal']

for i, (clf, clf_name) in enumerate(zip(classifiers, classifier_names), 1):

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    conf_mat = confusion_matrix(y_test, y_pred)
    
    plt.subplot(3, 2, i)
    sns.heatmap(conf_mat, annot=True, fmt='g', cmap='Blues',xticklabels=labels, yticklabels=labels)
    plt.title(f'Confusion Matrix: {clf_name}')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')

plt.tight_layout()
plt.show()
```
